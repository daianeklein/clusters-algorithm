# KMeans

### Importando libs

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

### Criando dados

```python
x = [20,  27,  21,  37,  46, 53, 55,  47,  52,  32,  39,  41,  39,  48,  48]  
y = [1000,1200,2900,1850,900,950,2000,2100,3000,5900,4100,5100,7000,5000,6500]
```

```python
dados = np.array([[20,1000],[27,1200],[21,2900],[37,1850],[46,900],
                 [53,950],[55,2000],[47,2100],[52,3000],[32,5900],
                 [39,4100],[41,5100],[39,7000],[48,5000],[48,6500]])
```

### Feature Scaling

```python
scaler = StandardScaler()
dados = scaler.fit_transform(dados)
```

### KMeans

```python
## Definindo os parametros
kmeans = KMeans(n_clusters = 3)

## Treinando o modelo
kmeans.fit(dados)
```

```python
## Criando a variavle com a posicao dos centroids
centroids = kmeans.cluster_centers_

## Criando a variavel com os rotulos 
labels = kmeans.labels_
```

```python
## definindo as cores
cores = ['b.', 'r.', 'g.']

## Plotando os valores conforme as cores
for i in range(len((x))):
    plt.plot(dados[i][0], dados[i][1], cores[labels[i]], markersize = 15)
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'o')
```

## Make_blobs

### Importando as bibliotecas

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

### Clicando e plotando valores aleatorios

```python
## Criando 'pontos'
x, y = make_blobs(n_samples = 500, centers = 3)

## Plotando
plt.scatter(x[:, 0], x[:,1])
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(x)
```

```python
previsoes = kmeans.predict(x)
plt.scatter(x[:,0], x[:,1], c=previsoes)
```
